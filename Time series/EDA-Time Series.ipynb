```python
def rolling_forecast(df: pd.DataFrame, train_len: int, horizon: int, window: int, method: str) -> list:
    
    total_len = train_len + horizon
    
    if method == 'mean':
        pred_mean = []
        
        for i in range(train_len, total_len, window):
            mean = np.mean(df[:i].values)
            pred_mean.extend(mean for _ in range(window))

        return pred_mean

    elif method == 'last':
        pred_last_value = []
        
        for i in range(train_len, total_len, window):
            last_value = df[:i].iloc[-1].values[0]
            pred_last_value.extend(last_value for _ in range(window))
            
        return pred_last_value
    
    elif method == 'MA':
        pred_MA = []
        
        for i in range(train_len, total_len, window):
            model = SARIMAX(df[:i], order=(0,0,2))
            res = model.fit(disp=False)
            predictions = res.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean.iloc[-window:]
            pred_MA.extend(oos_pred)
            
        return pred_MA
```

```python
pred_df = test.copy()

TRAIN_LEN = len(train)
HORIZON = len(test)
WINDOW = 2

pred_df['pred_mean'] = rolling_forecast(df_diff,TRAIN_LEN,HORIZON,WINDOW,'mean')
pred_df['pred_last_value'] = rolling_forecast(df_diff,TRAIN_LEN,HORIZON,WINDOW,'last')
pred_df['pred_MA'] = rolling_forecast(df_diff,TRAIN_LEN,HORIZON,WINDOW,'MA')

pred_df.head()
```

# ARMA Model

```python
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
np.random.seed(42)
ar1 = np.array([1,-0.33])
ma1 = np.array([1,0.9])
ARMA_1_1 = ArmaProcess(ar1,ma1).generate_sample(nsample=1000)
```

```python
plt.plot(ARMA_1_1)
```

```python
# Performing ADFuller test
from statsmodels.tsa.stattools import adfuller
adfuller_test = adfuller(ARMA_1_1)
print(f'ADF Statistic: {adfuller_test[0]}')
print(f'p-value: {adfuller_test[1]}')
```

```python
# Plot ACF and PACF
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(ARMA_1_1)
plt.show()
```

```python
plot_pacf(ARMA_1_1)
plt.show()
```

```python
# Combination of P and Q values
from itertools import product
ps = range(0,4,1)
qs = range(0,4,1)
order_list = list(product(ps,qs))
```

```python
# Function to fit all the unique ARMA(p,q) models
from typing import Union
from tqdm import tqdm_notebook
from statsmodels.tsa.statespace.sarimax import SARIMAX

def optimize_ARMA (endog:Union[pd.Series,list],order_list:list)-> pd.DataFrame:
    results = []
    for order in tqdm_notebook(order_list):
        try:
            model = SARIMAX(endog,order=(order[0],0,order[1]),
                            simple_differencing=False).fit(disp=False)
        except:
            continue
            
        aic=model.aic
        results.append([order,aic])
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q)', 'AIC']
    
    # Sort in ascending order the list, lower AIC
    result_df.sort_values(by='AIC',ascending=True,inplace=True)
    return result_df
```

```python
optimize_ARMA(ARMA_1_1,order_list)
```

```python
model = SARIMAX(ARMA_1_1,order=(1,0,1),simple_differencing=False)
model_fit = model.fit(disp=True)
residuals = model_fit.resid

from statsmodels.graphics.gofplots import qqplot
qqplot(residuals,line='45')
plt.show()
```

```python
model_fit.plot_diagnostics(figsize=(10,8))
plt.show()
```

```python
# Ljung box test 
from statsmodels.stats.diagnostic import acorr_ljungbox
lbvalue, pvalue = acorr_ljungbox(residuals,np.arange(1,11,1))
print(pvalue)
```

```python
acorr_ljungbox(residuals,np.arange(1,11,1))
```

The resulting list of p-values shows that each is above 0.05. Therefore, at each lag, the
null hypothesis cannot be rejected, meaning that the residuals are independently distributed
and uncorrelated.

```python
from itertools import product
ps = range(0,13,1)
qs = range(0,13,1)
Ps = [0]
Qs = [0]
d = 2
D = 0
S = 12
ARIMA_order_list = list(product(ps,qs,Ps,Qs))
```

```python
train = df[:-12]
test = df[-12:]
print(len(train))
print(len(test))
```

```python
from typing import Union
from tqdm import tqdm_notebook
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings('ignore')
def optimize_SARIMA(endog:Union[pd.Series,list],order_list:list, d:int, D:int, s:int)-> pd.DataFrame:
    results = []
    for order in tqdm_notebook(order_list):
        try:
            model = SARIMAX(endog,
                           order=(order[0],d,order[1]),
                            seasonal_order=(order[2],D,order[3],s),
                            simple_differencing=False).fit(disp=False)
        except:
            continue
        aic = model.aic
        results.append([order,aic])
    result_df = pd.DataFrame(results)
    result_df.columns = ['p,q,P,Q','AIC']
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC',ascending=True).reset_index(drop=True)
    return result_df
```

```python
results_df = optimize_SARIMA(train,ARIMA_order_list,d,D,S)
results_df
```
