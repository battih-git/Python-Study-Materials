```python
# Simple Moving average - Window size 20
df_sma = df['Open'].rolling(20).mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,4))
df_sma.plot(legend=True,label="Moving average")
df['Open'].plot(legend=True)
```

```python
# Cumulative moving average
plt.figure(figsize=(12,4))
df['Open'].expanding().mean().plot(legend = True,label = 'Cumulative Moving average')
df['Open'].plot(legend=True)
```

```python
# Exponential Moving Average
df['Open'].ewm(alpha=0.05,adjust=False).mean().plot(legend=True,label='EWM')
df['Open'].plot(legend=True)
```
